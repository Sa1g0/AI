**Goal:**
Transforming raw data into a format understandable for machine learning algorithms (numbers only, no empty values).

```python
import pandas as pd
import numpy as np
```

```python
data_small = [
    {"emp_id": 101, "gender": "Male", "department": "IT", "salary": 80000, "hire_date": "2019-01-15", "experience_years": 5, "age": 29},
    {"emp_id": 102, "gender": "Female", "department": "HR", "salary": np.nan, "hire_date": "2021-06-01", "experience_years": 2, "age": 34},
    {"emp_id": 103, "gender": "Male", "department": "Sales", "salary": 60000, "hire_date": "2023-01-10", "experience_years": np.nan, "age": 999}, # Outlier Age
    {"emp_id": 104, "gender": "Female", "department": "IT", "salary": 82000, "hire_date": "2015-03-20", "experience_years": 9, "age": 40},
    {"emp_id": 105, "gender": "Other", "department": "Sales", "salary": 55000, "hire_date": "2022-11-05", "experience_years": 1, "age": "twenty"} # Wrong type
]
```

```python
df_small = pd.DataFrame(data_small)
print("--- SMALL DATASET ---")
display(df_small)
```

SLIDE 1 Data Cleaning

```python

# Check types - note that 'age' is Object because of "twenty"
print(df_small.info())

```

```python
# Fix 'age': Force numeric, errors='coerce' turns "twenty" into NaN
df_small['age'] = pd.to_numeric(df_small['age'], errors='coerce')
print("\nAfter type conversion:")
display(df_small)
```

SLIDE 2 IMPUTATION STRATEGIES

```python
# Salary: Fill with Median
median_salary = df_small['salary'].median()
df_small['salary_filled'] = df_small['salary'].fillna(median_salary)
df_small
```

```python
# Experience: Fill with 0
df_small['experience_filled'] = df_small['experience_years'].fillna(0)
df_small
```

SLIDE 3 - HANDLING OUTLIERS

```python
# Detect Outlier in Age (e.g., > 100)
outlier_mask = df_small['age'] > 100
print("Outliers detected:")
display(df_small[outlier_mask])

# Cap at 100
df_small.loc[df_small['age'] > 100, 'age'] = 100
display(df_small[['emp_id', 'age']])
```

SLIDE 4 - FEATURE ENGINEERING

```python
# Convert to datetime
df_small['hire_date'] = pd.to_datetime(df_small['hire_date'])
df_small
```

```python
# Calculate Tenure (Days since hire)
now = pd.Timestamp.now()
df_small['tenure_days'] = (now - df_small['hire_date']).dt.days
display(df_small[['hire_date', 'tenure_days']])
```

SLIDE 5 - CATEGORY ENCODING

```python
# One-Hot Encoding for Department
df_encoded = pd.get_dummies(df_small, columns=['department'], prefix='dept')
display(df_encoded[['emp_id', 'dept_HR', 'dept_IT', 'dept_Sales']])
```
